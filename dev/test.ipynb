```python
from celluloid import Camera
```

```python
all_gasdens = sim.load_allfields('gasdens')
```

```python
all_gasdens.size, all_gasdens.snapshots, all_gasdens.item('3')
```

```python
quees = all_gasdens.item('0')
```

```python
quees.keys()
```

```python
plt.ioff()
fig,ax = plt.subplots()

camara = Camera(fig)

for snapshot in all_gasdens.snapshots:
    gasdens = all_gasdens.item(str(snapshot)).item('gasdens')
    planet_disk,mesh = gasdens.meshslice(slice="z=0,ir='24:48',iphi='176:207'") # r = '0.8:1.2'
    ax.contour((mesh.x-1)*sim.UL/6371e5,mesh.y*sim.UL/6371e5,planet_disk*sim.URHO*1e3,levels=100)
    ax.set_xlabel("$x$ [au]")
    ax.set_ylabel("$y$ [au]")
    c.set_label("$\Sigma$ [kg/m$^2$]")
    ax.axis('equal')

    camara.snap()
plt.ion()
```

```python
animacion = camara.animate()
```
